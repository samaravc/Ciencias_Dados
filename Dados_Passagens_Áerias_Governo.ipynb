```python
#Importando a biblioteca
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Ler o arquivo e criar o DataFrame
df = pd.read_excel('201710_EmissaoPassagens_SCDP.xlsx', skiprows=3, skipfooter=3) 
```

```python
df.columns = ['Código do órgão superior', 'Nome do órgão superior', 'Código do órgão solicitante da viagem', 
              'Nome do órgão solicitante da viagem', 'N. PCDP', 'N. Reserva/Localizador', 'Data Emissão Bilhete',
              'Data Embarque', 'Valor Tarifa Comercial', 'Percentual Desconto Aplicado', 'Valor Tarifa Governo',
              'Valor Tarifa Embarque', 'Valor Bilhete', 'Companhia Aérea', 'Classe Tarifária Bilhete', 'Regra Tarifária',
              'No Show', 'Remarcado', 'Cancelado', 'Valor Multas', 'Valor Reembolso', 'Diferença de Tarifa',
              'Situação Final Bilhete']
```

```python
#Verificar as primeiras 5 linhas do DtaFrame
df.head()
```

```python
#Estatístca descritiva dos valores das tarifas/multas/reembolso/ bilhete em R$
df.describe()
```

```python
df['Companhia Aérea'].value_counts() 
```

```python
# Qual a compania áeria que mais vendeu passagens em 2017
# Utilizou-se a frequência Absoluta pra a variavel Compania Aéria
df['Companhia Aérea'].value_counts()
df['Companhia Aérea'].value_counts().plot.barh()
plt.xlabel('Compania Aérea')
plt.ylabel('Qtd de passagen')
```

```python
# Histograma da da istribuição de frequencia do valor do bilhete em R$
plt.hist(df['Valor Bilhete'], bins=12)
plt.xlabel('Valor do bilhete R$')
plt.ylabel('Frequencia absoluta')
plt.show()
```

```python
plt.hist(df, values="Valor Bilhete", color="Companhia Aérea", legend="xxx", bins=12)
plt.show()
```

```python
# Qual a situação do bilhete mais frequenquente no ano de 2017.
# Utilizou-se a frequência Absoluta pra a variavel Situação do bilhete.
df['Situação Final Bilhete'].value_counts()
df['Situação Final Bilhete'].value_counts().plot.barh()
plt.xlabel('Quantidade')
plt.ylabel('Situação Final do Bilhete')
```
